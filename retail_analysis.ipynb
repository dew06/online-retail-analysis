```python
# prompt: plot same chart but this time use percentage contribution.Show % on the bar

import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

# Assuming df4 is your DataFrame with 'Country' and 'TotalSales' columns

country_wise_sales = df4.groupby('Country')['TotalSales'].sum()
total_sales = country_wise_sales.sum()
percentage_contributions = (country_wise_sales / total_sales) * 100

top_countries = df4.groupby('Country')['TotalSales'].sum().sort_values(ascending=False).head(5)

ax = percentage_contributions.plot(kind='barh', title='Top 5 Countries by Total Sales (%)')
plt.gca().set_xlabel('Percentage Contribution of Total Sales')
plt.gca().set_ylabel('Country')
plt.grid(axis='x')

# Add percentage values to the bars
for i, v in enumerate(percentage_contributions):
    ax.text(v + 0.5, i, f"{v:.1f}%", color='black', va='center')

plt.show()
```

```python
# prompt: In the above chart show only the top 5 countries with their percentage

import matplotlib.pyplot as plt
top_countries_with_percentage = percentage_contributions[top_countries.index].sort_values(ascending=False)

ax = top_countries_with_percentage.plot(kind='barh', title='Top 5 Countries by Total Sales (%)')
plt.gca().set_xlabel('Percentage Contribution of Total Sales')
plt.gca().set_ylabel('Country')
plt.grid(axis='x')

# Add percentage values to the bars
for i, v in enumerate(top_countries_with_percentage):
    ax.text(v + 0.5, i, f"{v:.1f}%", color='black', va='center')

plt.show()
```

4.RFM Analysis

```python
current_dt = df4["InvoiceDate"].max() + pd.Timedelta(days=1)
current_dt
```

```python
rfm = df4.groupby("CustomerID").agg({
    "InvoiceDate": lambda x: (current_dt - x.max()).days,
    "InvoiceNo": "count",
    "TotalSales": "sum"
})
rfm.columns = ["Recency","Frequency","Monetary"]
rfm.head(5)
```

```python
#Segment customers based on RFM


rfm['R_Segment'] = pd.qcut(rfm['Recency'],4,labels=[4,3,2,1])
rfm['F_Segment'] = pd.qcut(rfm['Frequency'],4,labels=[1,2,3,4])
rfm['M_Segment'] = pd.qcut(rfm['Monetary'],4,labels=[1,2,3,4])
rfm['RFM_Score'] = rfm[['R_Segment', 'F_Segment', 'M_Segment']].sum(axis=1)

rfm.head()
```

```python
rfm.sort_values(by="RFM_Score",ascending=False)
```

##### 5.Customer Churn Analysis

```python
df4.head()
```

```python
customer_last_purchase = df4.groupby("CustomerID")["InvoiceDate"].max()
```

```python
current_dt
```

```python
customer_last_purchase = (current_dt - customer_last_purchase).dt.days
customer_last_purchase.head()
```

```python
churn_threshold = 90
churn_customers = customer_last_purchase[customer_last_purchase > churn_threshold]
churn_customers
```

```python
print("No. of churn customers = ",len(churn_customers))
```

```python
plt.figure(figsize=(10, 6))
plt.hist(customer_last_purchase, bins=50, color='skyblue', alpha=0.7)
plt.axvline(churn_threshold, color='red', linestyle='dashed', linewidth=2)
plt.title('Customer Churn Distribution')
plt.xlabel('Days Since Last Purchase')
plt.ylabel('Number of Customers')
plt.show()
```
